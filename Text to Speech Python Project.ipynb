```python
from tkinter import *
from gtts import gTTS
import pygame
import os
from datetime import datetime

# Create the main Tkinter window
root = Tk()
root.geometry('350x300')
root.resizable(0, 0)
root.config(bg='ghost white')
root.title('Text To Speech')

# Create a label for the application title and pack it into the window
Label(root, text='Text To Speech (TTS)', font=('Helvetica', 20, 'bold'), bg='#1F618D', fg='white').pack()

# Create a label for the instruction and place it at specified coordinates
Label(root, text='Enter Text', font=('Helvetica', 15, 'bold'), bg='#1F618D', fg='white').place(x=20, y=60)

# Create a StringVar to store the text entered by the user
Msg = StringVar()

# Create an entry widget for text input and link it to the Msg variable
entry_field = Entry(root, textvariable=Msg, width='50')
entry_field.place(x=20, y=100)

# Set the filename for the temporary audio file
filename = 'speech.mp3'

# Define a function to convert text to speech
def Text_to_speech():
    # Retrieve the text entered by the user
    Message = entry_field.get()

    # Check if pygame.mixer is already initialized and quit if true
    if pygame.mixer.get_init():
        pygame.mixer.quit()

    # Create a gTTS object with the entered text and save it as an MP3 file
    speech = gTTS(text=Message)
    speech.save(filename)

    # Initialize pygame.mixer, load the saved MP3 file, and play it
    pygame.mixer.init()
    pygame.mixer.music.load(filename)
    pygame.mixer.music.play()
    
    # Clear the content of the entry field after playing the speech
    entry_field.delete(0, 'end')
    
# Define a function to reset the text entry
def Reset():
    Msg.set("")

# Create PLAY button with a specified font, command, and position
Button(root, text="PLAY", font='arial 15 bold', command=Text_to_speech, width=4, bg='#4CAF50', fg='white').place(x=25, y=140)

# Create EXIT button with a specified font, command, background color, and position
Button(root, text='EXIT', font='arial 15 bold', command=root.destroy, bg='#E74C3C', fg='white').place(x=100, y=140)

# Create RESET button with a specified font, command, and position
Button(root, text='RESET', font='arial 15 bold', command=Reset, bg='#3498DB', fg='white').place(x=175, y=140)

# Run the main event loop to display the GUI
root.mainloop()
```
